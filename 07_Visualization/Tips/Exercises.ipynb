### Step 1. Import the necessary libraries:

```python
import pandas as pd
import numpy as np

# visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns

# set seaborn style to white
sns.set_style("white")
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv('tips.csv')
tips.head()
```

### Step 4. Delete the Unnamed 0 column

```python
del tips['Unnamed: 0']
tips.head()
```

```python
#Orginizing data
total_bill = tips['total_bill']
tip = tips['tip']
sex = tips['sex']
smoker = tips['smoker']
day = tips['day']
time = tips['time']
size = tips['size']
```

### Step 5. Plot the total_bill column histogram

```python
# create histogram
plt.hist(total_bill)

#labeling x and y axis
plt.xlabel('Value')#label x
plt.ylabel('Frequency')#label y
plt.title('Total Bill') #adding title of the graph

sns.despine()# take out the right and upper borders

#second one looks much better
```

```python
# create histogram
ttbill = sns.distplot(tips.total_bill);

#labeling x and y axis
plt.xlabel('Value')#label x
plt.ylabel('Frequency')#label y
plt.title('Total Bill') #adding title of the graph

# take out the right and upper borders
sns.despine()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.scatter(total_bill,tip)
#labeling x and y axis
plt.xlabel('Total Bill')#label x
plt.ylabel('Tip')#label y
plt.title('Total Bill vs Tip') #adding title of the graph
plt.show()
```

```python
#this way combines a scatter plot and histogram.
#like the first way better, not as much noise.
sns.jointplot(x = 'total_bill', y = 'tip', data = tips)
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.pairplot(tips)
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.scatter(day, total_bill, color='red')
#labeling x and y axis
plt.xlabel('Day')#label x
plt.ylabel('Total Bill')#label y
plt.title('Day vs Total Bill') #adding title of the graph
```

```python
sns.stripplot(x = "day", y = "total_bill", data = tips, jitter = True)
#jitter is usefull when there is a lot of points, prevents as much overlapp
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.stripplot(x = "tip", y = "day", hue = 'sex', data = tips, jitter = True)
#Hue used for seperating 
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(x= 'day', y = 'total_bill', hue = 'time', data = tips)
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
#seborn style
sns.set(style = 'ticks')

#create facet grid
g = sns.FacetGrid(tips, col = 'time')
g.map(plt.hist, 'tip')
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
g2 = sns.FacetGrid(tips, col = "sex", hue = "smoker")
g2.map(plt.scatter, "total_bill", "tip", alpha =.7)

g2.add_legend();
```
